# **Reading Web-sites** #

```python
import requests
from bs4 import BeautifulSoup

def reading_data():
    url = 'https://www.tableau.com/learn/articles/blogs-about-machine-learning-artificial-intelligence'

    response = requests.get(url)

    if response.status_code == 200:
        print("Website opened successfully\n")

        soup = BeautifulSoup(response.text, 'html.parser')

        section_id_prefix = "paragraph-id--"
        section_index = 195062

        while True:
            section_id = f"{section_id_prefix}{section_index}"
            section = soup.find("section", {"id": section_id})

            if not section:
                break

            heading_tag = section.find("h3")
            if heading_tag:
                print("Heading:", heading_tag.text.strip())

            subheading_tag = section.find("p")
            if subheading_tag:
                print("Content:", subheading_tag.text.strip())

            print("-" * 50)

            section_index += 1
    else:
        print(f"Unable to fetch the webpage. Status code: {response.status_code}")

reading_data()
```

# **Exception Handling** #

**Error:** Errors are serious issues that a program should not try to handle. They are usually problems in the code’s logic or configuration and need to be fixed by the programmer. Examples include syntax errors and memory errors.

**Exception:** Exceptions are less severe than errors and can be handled by the program. They occur due to situations like invalid input, missing files or network issues.

```python
import logging
logging.basicConfig(filename='app.log', level=logging.DEBUG)


def read_file(file_path):
    try:
        with open(file_path, 'r') as file:
            content = file.read()
        return content
    except FileNotFoundError as e:
        logging.error(f"FileNotFoundError: The file at {file_path} was not found.")
        print(f"Error: The file at {file_path} was not found.")

    except PermissionError as e:
        logging.error(f"PermissionError: No permission to access the file at {file_path}.")
        print(f"Error: No permission to access the file at {file_path}.")

    except Exception as e:
        logging.error(f"An unexpected error occurred: {e}")
        print("An unexpected error occurred.")
    return None


def convert_to_integer(value):
    try:
        return int(value)
    except ValueError as e:
        logging.error(f"ValueError: Unable to convert '{value}' to an integer.")
        print(f"Error: '{value}' is not a valid integer.")

    except Exception as e:
        logging.error(f"An unexpected error occurred: {e}")
        print("An unexpected error occurred.")
    return None

file_content = read_file('nonexistent_file.txt')
if file_content:
    print(file_content)

user_input = 'abc'
number = convert_to_integer(user_input)
if number is not None:
    print(f"Converted number: {number}")
```
