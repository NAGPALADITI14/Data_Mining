# **Matplotlib**

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 4]

hist_data = np.random.randn(1000)

pie_sizes = [30, 20, 45, 5]
labels = ['A', 'B', 'C', 'D']

x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)

box_data = [np.random.randn(100) for _ in range(4)]

fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Simple Plot
axs[0, 0].plot(x, y, label='sin(x)', color='blue')
axs[0, 0].set_title('Simple Plot')
axs[0, 0].legend()

# Bar Plot
axs[0, 1].bar(categories, values, color='green')
axs[0, 1].set_title('Bar Plot')

# Histogram
axs[0, 2].hist(hist_data, bins=10, color='purple', alpha=0.7)
axs[0, 2].set_title('Histogram')

# Pie Chart
axs[1, 0].pie(pie_sizes, labels=labels, autopct='%1.1f%%', colors=['pink', 'blue', 'purple', 'green'])
axs[1, 0].set_title('Pie Chart')

# Scatter Plot
axs[1, 1].scatter(x_scatter, y_scatter, color='black')
axs[1, 1].set_title('Scatter Plot')

# Box Plot
axs[1, 2].boxplot(box_data)
axs[1, 2].set_title('Box Plot')

plt.tight_layout()
plt.show()
```

# **Seaborn**

```python
# Matplot implementation

import numpy as np
from matplotlib import pyplot as plt
def sinplot(flip=2):
  x = np.linspace(1,10,100)
  for i in range(1,7):
    plt.plot(x,np.sin(x+i*.5)*(6-i)*flip)

sinplot()
plt.show()

# seaborn implementation
import seaborn as sb
sb.set()
sinplot()
plt.show()
```

```python
# Figure styles
sb.set_style("whitegrid")
sinplot()
plt.show()

sb.set_style("dark")
sinplot()
plt.show()

sb.set_style("darkgrid")
sinplot()
plt.show()

sb.set_style("ticks")
sinplot()
plt.show()

sb.set_style("white")
sinplot()
plt.show()
```

```python
# Remove axes spines
sb.set_style("white")
sinplot()
sb.despine()
plt.show()

# overriding the elements
sb.axes_style()

sb.set_style("white",{'grid.linestyle' : '--' })
sinplot()
sb.despine()
plt.show()
```

```python
# Color Palette

palette=sb.color_palette()
sb.palplot(palette)
plt.show()

sb.palplot(sb.color_palette("Greens"))
plt.show()

sb.palplot(sb.color_palette("Purples"))
plt.show()

sb.palplot(sb.color_palette("pink"))
plt.show()

# Diverging Color Palette
sb.palplot(sb.color_palette("BrBG", 7))
plt.show()

# setting default color palette
sb.set_style("white")
sb.set_palette("husl")
sinplot()
plt.show()
```

```python
# histograms

df = sb.load_dataset('iris')
sb.displot(df['petal_length'],kde=False)
plt.show()

# kernel density estimates(kde)
# Kernel Density Estimation (KDE) is a way to estimate the probability density function of a continuous random variable. It is used for non-parametric analysis.
sb.displot(df['petal_length'],kind="kde")
plt.show()

# distplot() is used to visualize the parametric distribution of a dataset.
sb.displot(df['petal_length'],kde=True)
plt.show()

# Scatter Plot
sb.jointplot(x = 'petal_length',y = 'petal_width',data = df)
plt.show()

# hexbin Plot
# Hexagonal binning is used in bivariate data analysis when the data is sparse in density i.e., when the data is very scattered and difficult to analyze through scatterplots.
sb.jointplot(x='petal_length',y='petal_width',data=df,kind='hex')
plt.show()
```

```python
sb.set_style("ticks")
sb.pairplot(df, hue='species',diag_kind="kde",kind="scatter",palette="husl")
plt.show()
```

```python
# categorical Scatter Plot

# Stripplot
df = sb.load_dataset('iris')
sb.stripplot(x = "species", y = "petal_length", data = df)
plt.show()

# Jitter adds some random noise to the data
df = sb.load_dataset('iris')
sb.stripplot(x = "species", y = "petal_length", data = df,jitter=True)
plt.show()

# SwarmPlot
df = sb.load_dataset('iris')
sb.swarmplot(x = "species", y = "petal_length", data = df)
plt.show()
```

```python
# Distribution of Observations

# Boxplot
sb.boxplot(x = "species", y = "petal_length", data = df)
plt.show()

# violinplot
sb.violinplot(x = "species", y = "petal_length", data = df)
plt.show()

sb.violinplot(x = "petal_width", y = "petal_length", hue="species",data = df)
plt.show()
```
